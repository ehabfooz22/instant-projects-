```python
import re
import nltk

nltk.download('stopwords')
#from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer
ps = LancasterStemmer()

#all_stopwrds = stopwords.words('english')
#all_stopwrds.remove('not')
```

```python
corpus=[]

for i in range(0,10314):
    tweet = re.sub('[^a-z A-Z]',' ',df["message to examine"][i])
    tweet = tweet.lower() # convert text to lower case
    tweet = tweet.split()# Splits at space
    tweet = [ps.stem(word) for word in tweet] #if not word in set(all_stopwrds)
    tweet = ' '.join(tweet) # join elements of text with space
    corpus.append(tweet)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=200)
```

```python
x = cv.fit_transform(corpus).toarray()
y = df["label (depression result)"].values
```

```python
len(x[0])
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
Kcl = KNeighborsClassifier()
Kcl.fit(x_train,y_train)
```

```python
k_NN_training_accuracy=Kcl.score(x_train,y_train)
print(k_NN_training_accuracy)
```

```python
k_NN_testing_accuracy=Kcl.score(x_test,y_test)
print(k_NN_testing_accuracy)
```
