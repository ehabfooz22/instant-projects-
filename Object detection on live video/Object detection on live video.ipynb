```python
import cv2 as cv
```

```python
cap= cv.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)
```

```python
className=[]
classFile='coco.names.csv'
with open(classFile,'rt')as f:
    className=f.read().rstrip('\n').split('\n')
```

```python
className
```

```python
configPath='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath='frozen_inference_graph.pb'
```

```python
net=cv.dnn_DetectionModel(weightsPath,configPath)
net.setInputSize(320,320)
net.setInputScale(1.0/127.5)
net.setInputMean((127.5,127.5,127.5))
net.setInputSwapRB(True)
```

```python
while True:
    success,img = cap.read()
    classIds,confs,bbox=net.detect(img,confThreshold=0.5)
    print(classIds,bbox)
    if(len(classIds)!=0):
         for classId,confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
                cv.rectangle(img,box,color=(0,255.0),thickness=2)
                cv.putText(img,className[classId-1].upper(),(box[0]+10,box[1]+30),
                          cv.FONT_HERSHEY_COMPLEX,1,(0.255,0),2)
                cv.putText(img,str(round(confidence*100,2)),(box[0]+200,box[1]+30),
                          cv.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
    cv.imshow("Object Detection",img)
    cv.waitKey(1)
```
